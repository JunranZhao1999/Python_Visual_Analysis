```python
import seaborn as sns
import pandas as pd

sns.set(font='STHeiti',style = 'whitegrid',palette = 'crest')

# 创建一个DataFrame
data = {'x轴数据': [1, 2, 3, 4, 5, 6, 7, 8],
        'y轴数据': [10, 20, 30, 40, 50, 60, 70, 80],
        '类别列': ['X', 'Y', 'X', 'Z', 'Z', 'Y', 'X', 'Y'],
        '分面列': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B']}
df = pd.DataFrame(data)

# 绘制分面网格的关系图
sns.relplot(x='x轴数据', y='y轴数据', hue='类别列', data=df, col='分面列', col_wrap=2, palette='Set1', kind='scatter', markers=['o', 's', 'D'])

# 显示图形
import matplotlib.pyplot as plt
plt.show()
```

#### 实战2 | 钻石数据散点图

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head(5)
```

```python
diamonds['clarity'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')

fig,ax = plt.subplots(figsize = (8,8))

sns.set(style = 'whitegrid')
# clarity_categories = ['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1']
# hue_order映射clarity_ranking
clarity_ranking = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
sns.scatterplot(x='carat',y='price',data=diamonds,
                size='depth',sizes = (1,10),linewidth=0, # sizes参数用于设定散点大小的区间范围，能够改变散点的对比度。
                hue='clarity',hue_order=clarity_ranking,
                palette="ch:r=-.2,d=.3_r", # 订制这种蓝色色调的编码
                ax=ax)
sns.despine(left = True,bottom = True)

ax.legend(loc = 'lower right',ncol = 2,frameon = False)
ax.set_ylabel('Diamond Price')
ax.set_xlabel('Diamond Carat') # 钻石克拉
plt.title('Diamond Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')

fig,ax = plt.subplots(figsize = (8,8))

sns.set(style = 'whitegrid')
# clarity_categories = ['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1']
# hue_order映射clarity_ranking
clarity_ranking = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
sns.scatterplot(x='carat',y='price',data=diamonds,
                size='depth',sizes = (0.1,100),linewidth=0, # sizes参数用于设定散点大小的区间范围，能够改变散点的对比度。
                hue='clarity',hue_order=clarity_ranking,
                palette="ch:r=-.2,d=.3_r", # 订制这种蓝色色调的编码
                ax=ax)
sns.despine(left = True,bottom = True)

ax.legend(loc = 'lower right',ncol = 2,frameon = False)
ax.set_ylabel('Diamond Price')
ax.set_xlabel('Diamond Carat') # 钻石克拉
plt.title('Diamond Price')
plt.show()
```
