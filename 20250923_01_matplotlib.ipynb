```python
# 绘制一个柱状图（颜色）

import matplotlib.pyplot as plt
import matplotlib

data = [5, 6, 7, 8]
labels = ["a", "b", "c", "d"]
colors = ["red","yellow",'blue','black']

plt.bar(range(len(data)), data, color=colors)
plt.xticks(range(len(data)),labels)
plt.xlabel("i am x")
plt.ylabel("i am y")
plt.title("i am title")
plt.show()
```

```python
# 绘制一个柱状图（显示数据）
import matplotlib.pyplot as plt
import matplotlib

data = [5,6,7,8]
labels = ['a','b','c','d']
colors = ['red','yellow','orange','beige']

plt.bar(range(len(data)),data,color = colors,width = 0.5)
plt.xticks(range(len(data)),labels)
plt.xlabel('i am x')
plt.ylabel('i am y')
plt.title('i am title')
for i in range(len(data)):
    plt.text(x=i-0.05,y=data[i]+0.10,s='%d'%data[i])

plt.show()
```

### 绘制饼图

```python
# 绘制一个基础饼图

import matplotlib.pyplot as plt
import numpy as np

x = np.array([40,25,25,10])
plt.rcParams['font.sans-serif'] = ['STHeiti']
plt.pie(x,labels=['C++','Java','Python','C'],colors=['darkred','darkblue','green','orange'])
plt.title('学生选课分布饼图')
plt.show()
```

```python
# 绘制一个饼图（分开效果的）
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager

label_list = ["第一部分", "第二部分", "第三部分"] 
size = [55, 35, 10] 
color = ["red", "green", "blue"] 
explode = [0, 0.05, 0]  

plt.figure(figsize = (20,8),dpi=100)
plt.rcParams['font.sans-serif'] = ['STHeiti']
patches, labels_text, patches_text = plt.pie(size,
                                 explode = explode,
                                 colors = color,
                                 labels = label_list,
                                 labeldistance = 1.1,
                                 autopct = '%.2f%%',
                                 shadow = True,
                                 startangle = 90,
                                 pctdistance = 0.6)

plt.legend(patches,labels,
          title = '学生选课分布饼图',
          loc = 'center left',
          bbox_to_anchor=(1,1),
          ncol = 3)

plt.show()
```

```python
# 绘制一个多重变量的折线图
from matplotlib import pyplot as plt

dev_x = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
dev_y = [1, 4, 9, 16, 7, 11, 23, 18]
plt.plot(dev_x, dev_y, label="Group1",color='orange',marker='.',linestyle='--')

py_dev_y =[0, 10, 20, 30, 40, 45, 50, 55]
plt.plot(dev_x, py_dev_y, label="Group2",color='green',marker='.',linestyle=':')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("X和Y的关系")
plt.rcParams['font.sans-serif'] = ['STHeiti'] 
plt.legend()
plt.grid(True)
plt.show()
```

### 绘制曲线图

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['STHeiti'] 
plt.rcParams['axes.unicode_minus'] = False

x = np.arange(1,15)

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_xlabel('num',fontsize = 14)

ax1.plot(x,np.exp(x),'b--')
ax1.set_ylabel('exp',fontsize = 14)

ax2 = ax1.twinx()
ax2.plot(x,np.log(x),'ro:')
ax2.set_ylabel('log',fontsize = 14)

fig.legend(labels = ('exp','log'),loc = 'upper right')
plt.show()
```

```python
# 绘制一组复杂曲线图

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3)
fig = plt.figure(figsize = (15,15))

ax1 = fig.add_subplot(221)
ax1.plot(x,np.exp(x),'c+',label = r'exp(x)')
ax1.plot(x, x ** 2,'g*',label = r'$x^2$')
ax1.set_title('pic 1')
ax1.set_xlim(-3,3)
ax1.legend(loc = 'upper left')
ax1.grid()

ax2 = fig.add_subplot(222)
ax2.plot(x,np.sin(x),'m-',label = r'sin(x)')
ax2.plot(x,np.cos(x),'rs',label = r'cos(x)')
ax2.set_title('pic 2')
ax2.legend(loc = 'upper left')
ax2.grid(color = 'b', ls = '-',lw = 0.35)

ax3 = fig.add_subplot(223)
ax3.plot(x,0.5*x-2,'kv',label = r'0.5*x-2')
ax3.plot(x,2*x+1,'g+',label = r'2*x+1')
ax3.set_title('pic 3')
ax3.legend(loc = 'upper left')
ax3.grid()

ax4 = fig.add_subplot(224)
ax4.plot(x,-x**2,'b-',label = r'$-x^2$')
ax4.plot(x,x**3,'g+',label = r'$x^3$')
ax4.set_title('pic 4')
ax4.legend(loc = 'upper left',frameon = False)
ax4.grid()

plt.show()
```

### Figure 相关
